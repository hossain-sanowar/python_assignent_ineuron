#Write a function to compute 5/0 and use try/except to catch the exceptions.

```python
def division(number):
    number=int(input()) 
    try: 
        number%0 
    except ZeroDivisionError: 
        print("Denominator has to be greater than 0")
```

```python
division('number')
```
